```python
import geopandas
import matplotlib
import matplotlib.pyplot as plt

folder = raw_input("Insira o endereco do arquivo geojson a partir do diretorio onde se encontra o arquivo .py ")
coluna = raw_input("Insira o nome da coluna onde se encontram os valores que devem ser classificados ")
mapa = geopandas.read_file(folder)
mapa.plot(column=coluna, cmap='YlOrRd', scheme='fisher_jenks', figsize=(16, 14), facecolor='w', edgecolor='k', legend = True)
```

```python
from matplotlib.pyplot import figure
import numpy as np
figure(figsize=(16,14))
data = mapa.loc[:,[coluna]]
#plt.hist(data, bins = 10)
maximo = np.max(data)
minimo = np.min(data)
print(minimo)
print(maximo)
plt.hist(data, bins=10)
plt.show()
```
